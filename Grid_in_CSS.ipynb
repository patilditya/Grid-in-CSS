Problem Statement

Explain the use of grid-auto-row and grid-auto-column using code examples.

--> ### Problem Statement:  
**Explain the use of `grid-auto-rows` and `grid-auto-columns` in CSS Grid with code examples.**

---

### Explanation:

CSS Grid Layout provides a powerful way to create two-dimensional layouts on the web. Two of the properties used in CSS Grid are `grid-auto-rows` and `grid-auto-columns`. These properties control the size of rows and columns that are automatically created when there are more grid items than the defined grid areas.

1. **`grid-auto-rows`**:  
   The `grid-auto-rows` property controls the size of rows that are automatically created when grid items are placed outside the defined rows in a CSS Grid container.

2. **`grid-auto-columns`**:  
   Similarly, the `grid-auto-columns` property controls the size of columns that are automatically created when grid items are placed outside the defined columns in a CSS Grid container.

Both of these properties apply to the implicit grid, i.e., the grid that is automatically generated when the content exceeds the explicitly defined rows or columns in a grid.

### Code Examples

---

### Example 1: Using `grid-auto-rows`

Here, we will define a 2x2 grid (2 rows and 2 columns). However, we will place an item in a third row, which will be created automatically, and we will use `grid-auto-rows` to define its size.

```html
<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8">
  <meta name="viewport" content="width=device-width, initial-scale=1.0">
  <title>Grid Auto Rows Example</title>
  <style>
    .grid-container {
      display: grid;
      grid-template-rows: 100px 100px;
      grid-template-columns: 100px 100px;
      grid-gap: 10px;
      grid-auto-rows: 150px; /* Automatically created rows will be 150px tall */
    }

    .grid-item {
      background-color: #8e44ad;
      color: white;
      display: flex;
      align-items: center;
      justify-content: center;
    }
  </style>
</head>
<body>
  <div class="grid-container">
    <div class="grid-item">Item 1</div>
    <div class="grid-item">Item 2</div>
    <div class="grid-item">Item 3</div> <!-- This item will create a new row -->
  </div>
</body>
</html>
```

**Explanation:**
- In this example, the grid has two defined rows, each 100px tall (`grid-template-rows: 100px 100px`).
- However, we add a third grid item (`Item 3`), which will automatically create a new row.
- The new row will have a height of `150px`, as defined by the `grid-auto-rows` property.

### Example 2: Using `grid-auto-columns`

Similarly, we can control the size of automatically created columns using `grid-auto-columns`. Below is an example where we define a 2x2 grid, but we place an item in the third column, which will be created automatically.

```html
<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8">
  <meta name="viewport" content="width=device-width, initial-scale=1.0">
  <title>Grid Auto Columns Example</title>
  <style>
    .grid-container {
      display: grid;
      grid-template-rows: 100px 100px;
      grid-template-columns: 100px 100px;
      grid-gap: 10px;
      grid-auto-columns: 200px; /* Automatically created columns will be 200px wide */
    }

    .grid-item {
      background-color: #e74c3c;
      color: white;
      display: flex;
      align-items: center;
      justify-content: center;
    }
  </style>
</head>
<body>
  <div class="grid-container">
    <div class="grid-item">Item 1</div>
    <div class="grid-item">Item 2</div>
    <div class="grid-item">Item 3</div> <!-- This item will create a new column -->
  </div>
</body>
</html>
```

**Explanation:**
- The grid has two defined columns, each 100px wide (`grid-template-columns: 100px 100px`).
- We place a third grid item (`Item 3`), which will automatically create a new column.
- The new column will have a width of `200px`, as defined by the `grid-auto-columns` property.

---

### When to Use `grid-auto-rows` and `grid-auto-columns`:

- Use `grid-auto-rows` when you expect your grid to generate additional rows, and you want to control the size of those rows.
- Use `grid-auto-columns` when you expect your grid to generate additional columns, and you want to control the size of those columns.

These properties are helpful when the number of items is dynamic or unknown, and they allow you to manage the grid layout more efficiently by controlling the size of implicitly created rows and columns.

---

### Summary:

- `grid-auto-rows`: Defines the size of automatically created rows when the grid items exceed the defined rows.
- `grid-auto-columns`: Defines the size of automatically created columns when the grid items exceed the defined columns.

Question 5 :Problem Statement

Explain the difference between justify-items and justify-self using code examples.


--> The properties `justify-items` and `justify-self` are used in CSS to control the alignment of items within a grid or flex container. Although both deal with alignment, they are used in different contexts and have different scopes of influence.

### 1. **`justify-items`**
The `justify-items` property is used to align items within a **grid container** along the **inline axis** (typically the horizontal axis). It affects **all** grid items unless specifically overridden by the `justify-self` property.

### 2. **`justify-self`**
The `justify-self` property is used to align a specific **grid item** along the inline axis (horizontal axis) within its grid cell. This property gives you more granular control, allowing you to align individual items differently from others.

---

### Code Example:

#### Using `justify-items`:

In the following example, we use `justify-items` to align all the grid items within the grid container to the **center**.

```html
<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8">
  <meta name="viewport" content="width=device-width, initial-scale=1.0">
  <title>justify-items Example</title>
  <style>
    .grid-container {
      display: grid;
      grid-template-columns: repeat(3, 100px);
      grid-gap: 10px;
      justify-items: center; /* Align all items to the center */
      height: 200px;
      background-color: lightgray;
    }

    .grid-item {
      background-color: lightblue;
      height: 50px;
      text-align: center;
      line-height: 50px;
    }
  </style>
</head>
<body>
  <div class="grid-container">
    <div class="grid-item">Item 1</div>
    <div class="grid-item">Item 2</div>
    <div class="grid-item">Item 3</div>
  </div>
</body>
</html>
```

In this example:
- `justify-items: center;` aligns **all grid items** to the center horizontally within their respective grid cells.

---

#### Using `justify-self`:

In the following example, we use `justify-self` to align **individual grid items** differently. The first item will be aligned to the start, the second to the center, and the third to the end.

```html
<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8">
  <meta name="viewport" content="width=device-width, initial-scale=1.0">
  <title>justify-self Example</title>
  <style>
    .grid-container {
      display: grid;
      grid-template-columns: repeat(3, 100px);
      grid-gap: 10px;
      height: 200px;
      background-color: lightgray;
    }

    .grid-item {
      background-color: lightblue;
      height: 50px;
      text-align: center;
      line-height: 50px;
    }

    .item-1 {
      justify-self: start; /* Align item 1 to the start */
    }
    
    .item-2 {
      justify-self: center; /* Align item 2 to the center */
    }
    
    .item-3 {
      justify-self: end; /* Align item 3 to the end */
    }
  </style>
</head>
<body>
  <div class="grid-container">
    <div class="grid-item item-1">Item 1</div>
    <div class="grid-item item-2">Item 2</div>
    <div class="grid-item item-3">Item 3</div>
  </div>
</body>
</html>
```

In this example:
- `justify-self: start;` aligns **Item 1** to the start of its grid cell.
- `justify-self: center;` aligns **Item 2** to the center of its grid cell.
- `justify-self: end;` aligns **Item 3** to the end of its grid cell.

---

### Key Differences:

- **Scope**:
  - `justify-items`: Affects **all grid items** within the container.
  - `justify-self`: Affects **a specific grid item** within its grid cell.

- **Use Case**:
  - `justify-items`: Use this when you want to apply the same alignment to all grid items inside the container.
  - `justify-self`: Use this when you want to apply a specific alignment to individual grid items.

These properties are part of the **CSS Grid** layout and do not affect **flexbox**.
